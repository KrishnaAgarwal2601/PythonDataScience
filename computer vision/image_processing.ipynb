```python
from PIL import Image, ImageDraw, ImageFont, ImageEnhance, ImageFilter
import requests # download data from url
```

```python
# local image
local_image_path = "D:\image.webp"
# image from an url
image_url = "https://static.toiimg.com/photo/102165926.cms"
```

```python
im1 = Image.open(local_image_path)
print(im1)
```

```python
def get_image_from_url(url):
    response = requests.get(url, stream= True)
    return Image.open(response.raw)
```

```python
img2 = get_image_from_url(image_url)
print(img2)
```

```python
# image attributes
print("Image 1")
print("Mode", im1.mode)
print("Format", im1.format)
print("Size", im1.size)
print("Width", im1.width)
print("Height", im1.height)
print("Info", im1.info)
```

```python
print("Image 1 metadata")
print(img2.getexif())
```

METHODS

```python
# resize image
im11 =im1.resize((400, 400), resample=Image.LANCZOS)
im11
```

```python
w, h = img2.size
img22 = img2.resize((w//3, h//3),resample=Image.LANCZOS)
# rotate image
img22.rotate(90,fillcolor='white', expand=True)
```

```python
# flip image
img22.transpose(Image.FLIP_LEFT_RIGHT)
```

```python
# crop image
print(im11.size)
im11.crop((0,0,im11.width,255))
```

```python
im11.crop((50,0,im11.width-170,im11.height-140))
```

```python
# creating a blank imageg
yimg = Image.new("RGB",(400,400),"yellow")
yimg
```

```python
# blending images - Rule: same size and size format
im11.mode
```

```python
Image.blend(im11,yimg,0.4)
```

```python
# paste image
rimg = Image.new("RGB",(400,400),"red")
rimg.paste(im11.resize((100,100)),(150,150))
rimg
```
