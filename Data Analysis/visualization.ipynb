```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,3*np.pi,50)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.figure(figsize=(12,7))
plt.plot(t,x,'o-')
plt.plot(t,y,'H-g', markersize=10 , markeredgecolor='black')
```

```python
years = [2010,2011,2012,2013,2014,2015]
prices = [120,90,140,150,157,148]
fig= plt.figure(figsize=(10,6))
ax1 = fig.add_axes([0,0,1,1]) #[x,y,width,height]
ax2 = fig.add_axes([0,0,0.2,0.2])
ax3 = fig.add_axes([0.8,0.8,0.2,0.2])
ax4 = fig.add_axes([0.8,0,0.2,0.2])
ax4 = fig.add_axes([0.4,0.4,0.3,0.3])
```

```python
fig = plt.figure(figsize=(6,4))
ax1 = fig.add_axes([0,0,1,1])
ax1.bar(years,prices)
```

```python
plt.pie(prices,labels=years,autopct="%.2f%%")
plt.show()
```

### Customization

```python
plt.figure(figsize=(10,6),facecolor="orange")
plt.plot(years,prices,'rD-')
plt.title("Prices of Stocks")
plt.xlabel("Years")
plt.ylabel("Prices")
plt.hlines([140,140,150],
           [2010,2012,2010],
           [2012,2015,2013],
           linestyles=["dashed","solid"])
plt.vlines([2012,2013],
           [min(prices),min(prices)],
           [140,150],
           linestyles=['dotted','solid'],
           colors=['red','blue'])
plt.text(2011, 144,'Important year',fontdict={'size':15},color='green')
plt.text(2012.5, 130,'Increase \nin price',fontdict={'size':10},color='green')
plt.show()
```
